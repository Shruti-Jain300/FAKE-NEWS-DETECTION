```python
df_news = pd.concat([new_fake, new_true])
```

```python
df_news['text'] = df_news['title']+df_news['text']
df_news.drop(['title'],axis=1, inplace=True)
```

```python
df_news
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Concatenate all texts into a single string
all_text = ' '.join(df_news['text'].astype(str))

# Generate and plot the word cloud
wordcloud = WordCloud(width=800, height=400, random_state=21, max_font_size=110).generate(all_text)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
